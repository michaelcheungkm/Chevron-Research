```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import pandas as pd

options = Options()
options.headless = True
options.add_argument("--window-size=1920,1200")

df = pd.read_csv("keyword_links/glucose_links.csv")
url_list = df["search_url"].tolist()
article_list = []

for url in url_list:
    DRIVER_PATH = 'path/chromedriver'
    driver = webdriver.Chrome(options=options, executable_path=DRIVER_PATH)
    driver.get(url)
    
    title_span = driver.find_element(By.CLASS_NAME, "heading-title")
    title = title_span.text
    
    try:
        abstract_span = driver.find_element(By.CLASS_NAME, "selected")
        abstract = abstract_span.text
    except Exception as e:
        print(e)
    
    authors = []
    name_wrappers = driver.find_elements(By.CLASS_NAME, "full-name")
    for name in name_wrappers:
        authors.append(name.text)
    authors = list(set(authors))
    try:
        authors.remove("")
    except:
        pass
    
    article = {
        "link": url,
        "title": title,
        "abstract": abstract,
        "authors": authors
    }
    
    article_list.append(article)

article_list
```

```python
df_out = pd.DataFrame()
for article in article_list:
    df_out = df_out.append(article, ignore_index=True)
    
df_out
```

```python
search_term = "glucose"
out_name = "scraped_links/" + search_term + "_scraped.csv"
df_out.to_csv(out_name)
```
