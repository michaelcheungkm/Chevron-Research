```python
from sklearn.impute import KNNImputer

def perform_knn(df):
    imputer = KNNImputer(n_neighbors=5)
    return pd.DataFrame(imputer.fit_transform(df),columns = df.columns)

df_b_knn = perform_knn(df_b)
df_b_knn
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

def return_col_model(col, df):
    x = df.drop(columns=col)
    y = df[col]
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 100)
    model = LinearRegression()
    model.fit(x_train, y_train)
    return model

def cross_regression(df1, df2):
    df1_out = pd.DataFrame()
    df2_out = pd.DataFrame()
    common = df1.columns & df2.columns
    excl_1 = list(set(df1.columns) - set(common))
    excl_2 = list(set(df2.columns) - set(common))
    df1_out = df1.copy()
    df2_out = df2.copy()
    for col in excl_1:
        df_temp = df1[common]
        df_temp[col] = df1[col]
        df2_out[col] = return_col_model(col, df_temp).predict(df2[common])
    for col in excl_2:
        df_temp = df2[common]
        df_temp[col] = df2[col]
        df1_out[col] = return_col_model(col, df_temp).predict(df1[common])
    return df1_out, df2_out
        
def merge_dfs(df1, df2):
    merge_return = cross_regression(df1, df2)
    return merge_return[0].append(merge_return[1])

df_merge = merge_dfs(df_a, df_b_knn)
```

```python
common_with_chol = list(dftest_2.columns & dftest_1.columns)
common_with_chol.append('chol')
common = list(dftest_2.columns & dftest_1.columns)
def regress_column_numerical(col, df):
    x = df.drop(columns=col)
    y = df[col]
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 100)
    model = LinearRegression()
    model.fit(x_train, y_train)
    return model

def adjust(col, df1, df2):
    return df1[col].mean() - df2[col].mean()
```

```python
def return_col_numerical(col, df_with, df_without):
    df_temp = pd.DataFrame()
    common = list(df_with.columns & df_without.columns)
    common_with_col = list(df_with.columns & df_without.columns)
    common_with_col.append(col)
    df_without_common = df_without[common]
    df_with_common_with_col = df_with[common_with_col]
    df_temp[col] = regress_column_numerical(col, df_with_common_with_col).predict(df_without_common)
    df_temp[col] = df_temp[col].sub(adjust(col, df_temp, df_with))
    return df_temp[col]

dftest_12 = dftest_1.copy()
dftest_12["chol"] = return_col_numerical("chol", dftest_2, dftest_12)
dftest_12
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
def regress_column_class(col, df):
    x = df.drop(columns=col)
    y = df[col]
    x_train, x_test, y_train, y_test = train_test_split(x, y)
    model = LogisticRegression()
    model.fit(x_train, y_train)
    return model

def return_col_class(col, df_with, df_without):
    df_temp = pd.DataFrame()
    common = list(df_with.columns & df_without.columns)
    common_with_col = list(df_with.columns & df_without.columns)
    common_with_col.append(col)
    df_without_common = df_without[common]
    df_with_common_with_col = df_with[common_with_col]
    df_temp[col] = regress_column_class(col, df_with_common_with_col).predict(df_without_common)
    return df_temp[col]

dftest_13 = dftest_1.copy()
dftest_13["num"] = return_col_class("num", dftest_2, dftest_13)
np.array(dftest_13["num"])
```
